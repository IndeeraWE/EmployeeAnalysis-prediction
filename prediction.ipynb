```python
import pandas as pd
from   sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
train= pd.read_csv("C:\\Users\\HP\\Documents\\aug_train.csv")
train.head()
```

```python
new_df=train.fillna({
    'company_size':"50-99",
    'enrolled_university':"no_enrollment",
    'gender	':'Female',
    "company_type":"Pvt Ltd"
    
})
new_df.head()
```

```python
new_df12=new_df.replace({
    "never":4
    
})
```

```python
dummies=pd.get_dummies(new_df12.education_level)
dummies
```

```python
new_df12["education_level"].value_counts()
```

```python
df_123=new_df12[["training_hours","target"]].copy()
```

```python
df_123
```

```python
merged=pd.concat([df_123,dummies],axis="columns")
merged
```

```python
final=merged.drop(["Graduate"],axis="columns")
final
```

```python
x=final.drop(["target"],axis="columns")
x
```

```python
y=final.target
y
```

```python
 x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
 x_train
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_test,y_test)
```
